```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
feature=["id", "date", "yr_built","yr_renovated"]


#training
d_train=pd.read_csv("wk3_kc_house_train_data.csv",parse_dates=['date'])
d_train['zipcode'] = d_train['zipcode'].apply(str)
d_t=pd.get_dummies(d_train,drop_first=True)
d_t.sqft_basement[d_t.sqft_basement == 0] = 0
d_t.sqft_basement[d_t.sqft_basement != 0] = 1
group_base=d_t.groupby('sqft_basement').size()
print(group_base)
d_t.drop(feature, inplace = True,axis = 1)
x_train = d_t.drop(['price'], axis = 1)
y_train = d_t[['price']]
reg= linear_model.LinearRegression()
model=reg.fit(x_train,y_train)

#validation
d_valid=pd.read_csv("wk3_kc_house_valid_data.csv",parse_dates=['date'])
d_valid['zipcode'] = d_valid['zipcode'].apply(str)
d_v=pd.get_dummies(d_valid,drop_first=True)
d_v.sqft_basement[d_v.sqft_basement == 0] = 0
d_v.sqft_basement[d_v.sqft_basement != 0] = 1
d_v.drop(feature, inplace = True,axis = 1)
x_valid = d_v.drop(['price'], axis = 1)
y_valid = d_v[['price']]
scores = cross_val_score(model,x_valid,y_valid, cv=6)
print ('Cross-validated scores:', scores)
predictions = cross_val_predict(model,x_valid,y_valid, cv=6)
accuracy = metrics.r2_score(y_valid, predictions)


#testing

d_test=pd.read_csv("wk3_kc_house_test_data.csv",parse_dates=['date'])
d_test['zipcode'] = d_test['zipcode'].apply(str)
d_te=pd.get_dummies(d_test,drop_first=True)
d_te.sqft_basement[d_te.sqft_basement == 0] = 0
d_te.sqft_basement[d_te.sqft_basement != 0] = 1
d_te.drop(feature, inplace = True,axis = 1)
x_test = d_te.drop(['price'], axis = 1)
y_test = d_te[['price']]
pred = reg.predict(x_test)

print('Model_1')
print('\n\nDetails')
print('\nTraining')
r2=round(reg.score(x_train, y_train), 3)
print('R-squared (training) ',r2 )
adj_r2 = (1 - (1 - r2) * ((x_train.shape[0] - 1)/(x_train.shape[0] - x_train.shape[1] - 1)))
print('Adjusted R-squared (training) ',adj_r2)
print('\n\nValidation')
print ('Cross-Predicted Accuracy(R2):', accuracy)
print(f'Root Mean Squared Error: {np.sqrt(metrics.mean_squared_error(y_valid, predictions))}')
print(f'Mean absolute Percentage Error: {(100/len(y_valid)) * np.sum(np.absolute(y_valid - predictions)/y_valid)}')
print('\n\nTesting')

r2=round(reg.score(x_test,y_test), 3)
print('R-squared (testing)',r2)
adj_r2 = (1 - (1 - r2) * ((x_test.shape[0] - 1)/(x_test.shape[0] - x_test.shape[1] - 1)))
print('Adjusted R-squared (training) ',adj_r2)
print(f'Root Mean Squared Error: {np.sqrt(metrics.mean_squared_error(y_test,pred))}')
print(f'Mean absolute Percentage Error: {(100/len(y_test)) * np.sum(np.absolute(y_test - pred)/y_test)}')
```
